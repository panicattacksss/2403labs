```python
class Employee:
    def __init__(self, name, emp_id):
        self.name = name
        self.emp_id = emp_id

    def get_info(self):
        return f"Имя сотрудника: {self.name}, ID: {self.emp_id}"
```

```python
class Manager(Employee):
    def __init__(self, name, emp_id, department):
0        self.department = department

    def manage_project(self):
        return f"Менеджер {self.name} управляет проектами в отделе {self.department}."

    def get_info(self):
        return super().get_info() + f", Отдел: {self.department}"
```

```python
class Technician(Employee):
    def __init__(self, name, emp_id, specialization):
        super().__init__(name, emp_id)
        self.specialization = specialization

    def perform_maintenance(self):
        return f"Техник {self.name} выполняет техническое обслуживание по специализации {self.specialization}."

    def get_info(self):
        return super().get_info() + f", Специализация: {self.specialization}"
```

```python
class TechManager(Manager, Technician):
    def __init__(self, name, emp_id, department, specialization):
        Employee.__init__(self, name, emp_id)
        self.department = department
        self.specialization = specialization
        self.team = []

    def add_employee(self, employee):
        self.team.append(employee)

    def get_team_info(self):
        if not self.team:
            return f"У TechManager {self.name} нет членов команды."
        team_info = [employee.get_info() for employee in self.team]
        return f"Команда TechManager {self.name}:\n" + "\n".join(team_info)

    def get_info(self):
        return (
            super(Manager, self).get_info() +
            f", Отдел: {self.department}, Специализация: {self.specialization}"
        )
```

```python
emp1 = Employee("Рыбка", 4)
print(emp1.get_info())

emp2 = Employee("Малой", 5)
print(emp2.get_info())

manager = Manager("Авокадо", 3, "Тиктокер")
print(manager.get_info())
print(manager.manage_project())

technician = Technician("Максюшка Солдатов", 2, "Крутой")
print(technician.get_info())
print(technician.perform_maintenance())

tech_manager = TechManager("Дмитрий Солдатов", 1, "Бизнес", "Магазин одежды")
print(tech_manager.get_info())
print(tech_manager.manage_project())
print(tech_manager.perform_maintenance())

tech_manager.add_employee(manager)
tech_manager.add_employee(technician)
tech_manager.add_employee(emp1)
tech_manager.add_employee(emp2)


print(tech_manager.get_team_info())
```
