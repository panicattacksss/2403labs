```python
import math
import datetime
```

```python
while True:
    try:
        number = float(input("Введите положительное число: "))
        if number < 0:
            print("Корня отрицательного числа не существует")
        else:
            sqrt_value = math.sqrt(number)
            print(f"Квадратный корень: {sqrt_value}")
            print(f"Текущая дата и время: {datetime.datetime.now()}")
            break
    except ValueError:
        print("Введено некорректное значение. Попробуйте снова.")
```
