```python
def read_file(filename, mode='all'):
    try:
        with open(filename, 'r') as file:
            if mode == 'all':
                content = file.read()
                print("Содержимое файла:")
                print(content)
            elif mode == 'line':
                for line in file:
                    print(line, end='')
            else:
                print("Неизвестный режим чтения.")
    except FileNotFoundError:
        print(f"Ошибка: файл {filename} не найден.")

file_path = '/content/drive/My Drive/example.txt'

read_file(file_path, mode='all')
```

```python
def write_to_file():
    user_input = input("Введите текст: ")
    with open('user_input.txt', 'w') as file:
        file.write(user_input)
    print("Текст записан в файл user_input.txt")

write_to_file()
```

```python
def append_to_file():
    user_input = input("Введите текст для добавления в файл: ")
    with open('user_input.txt', 'a') as file:
        file.write("\n" + user_input)
    print("Текст добавлен в файл user_input.txt.")

append_to_file()
```

```python
def safe_read_file(filename, mode='all'):
    try:
        with open(filename, 'r') as file:
            if mode == 'all':
                content = file.read()
                print("Содержимое файла:")
                print(content)
            elif mode == 'line':
                for line in file:
                    print(line, end='')
            else:
                print(f"{mode} - неизвестный режим чтения.")
    except FileNotFoundError:
        print(f"Ошибка: файл {filename} не найден.")

safe_read_file('nonexistent.txt', mode='all')
```
