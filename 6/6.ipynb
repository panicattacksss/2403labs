```python
class UserAccount():
    def __init__(self, username, email, password):
        self.username = username
        self.email = email
        self.__password = password
    def set_password(self, new_password):
        self.__password = new_password
        print("пароль успешно изменён")
    def check_password(self, password):
        return self.__password == password
```

```python
user = UserAccount("MegaSigma228", "megasigma@ohio.com", "firstpass")
```

```python
print(user.check_password("firstpass"))
```

```python
print(user.check_password("firstpass1"))
```

```python
user.set_password("secondpass")
```

```python
print(user.check_password("firstpass"))
```

```python
print(user.check_password("secondpass"))
```

```python
class Vehicle():
    def __init__(self, make, model):
        self.make = make
        self.model = model
    def get_info(self):
        return f'{self.make} {self.model}'
```

```python
class Car(Vehicle):
    def __init__(self, make, model, fuel):
        super().__init__(make, model)
        self.fuel = fuel
    def get_info(self):
        return f"{self.make} {self.model}, fuel type: {self.fuel}"
```

```python
vehicle = Vehicle("Brand", "Car")
```

```python
car = Car("Nobrand", "Nocar", "Petrol")
```

```python
vehicle.get_info()
```

```python
car.get_info()
```
